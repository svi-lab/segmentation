# Characterization interconnected patterns

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

Let us import an image and make it binary.

Step 1 : denoising

```python
from skimage import io, exposure, restoration, img_as_ubyte

file_name =r".\data\spinodal_test_SVI.tif"
im_scale = io.imread(file_name, as_grey=True)
ima = img_as_ubyte(im_scale[20:940])
im = exposure.equalize_adapthist(ima)
im_tv = restoration.denoise_tv_bregman(im, 0.1)
```

```python
fig, axs = plt.subplots(1,2, figsize=(15,5))
ax1,ax2=axs
ax1.imshow(ima, cmap='gray')
ax1.title.set_text('raw image')
ax2.imshow(im_tv, cmap='gray')
ax2.title.set_text('after denoising')
```

Step 2: process segmentation, by Otsu tresholding.

Step 3: add labels.

```python
from skimage import segmentation, measure, morphology, draw, filters
from skimage.measure import label
from skimage.color import label2rgb

im_bin = im_tv > filters.threshold_otsu(im_tv)
im_bin = morphology.remove_small_objects(im_bin, 10)
im_bin_perco = morphology.remove_small_objects(im_bin, 10000)
im_bin_drop = np.logical_xor(im_bin, im_bin_perco)
im_label = measure.label(im_bin)
im_label_perco = measure.label(im_bin_perco)
im_lab = label2rgb(im_label, im_tv)
```

```python
fig, axs = plt.subplots(1,2, figsize=(15,5))
ax1,ax2=axs
ax1.imshow(im_bin, cmap='gray')
ax1.title.set_text('binary image')
ax2.imshow(im_lab)
ax2.title.set_text('labeled image')
```

A fisrt typical lengthscale can be obtained as the ratio of area/perimeter.

```python
properties=measure.regionprops(im_label)
area = []
perimeter = []
for prop in properties:
    area.append(prop.area)
    perimeter.append(prop.perimeter)
lenghtscale = np.asarray(area)/np.asarray(perimeter)

properties_perco=measure.regionprops(im_label_perco)
area_perco = []
perimeter_perco = []
for prop in properties_perco:
    area_perco.append(prop.area)
    perimeter_perco.append(prop.perimeter)
lenghtscale_perco = np.asarray(area_perco)/np.asarray(perimeter_perco)

plt.xlabel('A/P (pxl)')
plt.ylabel('number of objects')
plt.title('Lengthscale A/P')
n,bins, patch =plt.hist(lenghtscale, bins='auto', label='disconnected')
_=plt.hist(lenghtscale_perco, bins=bins, color='orange', label='percolating')
plt.legend();
```

### > So, we can distinguish the percolating phase (caractristic lenght of 30px) and the disconnected phase (15px)

We can think about a second typical length as the width of the wormlike part. It can be studied by skeletonization of the connected phase, and measure of the minimal distance to edges.

```python
skel = morphology.skeletonize(im_bin)
skel_perco = morphology.skeletonize(im_bin_perco)

plt.title('skeletonize')
plt.imshow(im_bin)
plt.contour(skel, colors='r');
```

```python
from scipy import ndimage
mat_skel=ndimage.distance_transform_edt(im_bin)
```

```python
d_skel = []
d_skel_perco = []
d_skel_thin = []
for i in range (0, np.shape(skel)[0]):
    for j in range (0, np.shape(skel)[1]):
        if (skel[i][j]==True):
            d_skel.append(0.5+mat_skel[i][j])
        if (skel_perco[i][j]==True):
            d_skel_perco.append(0.5+mat_skel[i][j])
```

```python
bins=50
plt.xlabel('local half-width')
plt.ylabel('number of pixels')
plt.title('Lengthscale from skeletonize')
_ =plt.hist(d_skel, bins=bins, label='disconnected')
_ =plt.hist(d_skel_perco, bins=bins, color='orange', label='percolating')
plt.legend();
```

However, this typical scale under estimate te disconnected objects (because few skelet pixels inside). 
Therefore, we can think about another lengthscale and try to compute the granulometry of the image. 
The granulometry is a local measure of the largest disk containing a given pixel and fully contained in the connected phase.

Let us first define a function for computing the granulometry.

```python
from joblib import Parallel, delayed
from scipy import ndimage

def _disk(n):
    struct = np.zeros((2 * n + 1,) * 2, dtype=np.bool)
    x, y = np.ogrid[:2 * n + 1, :2 * n + 1] 
    mask = (x - n)**2 + (y - n)**2 <= n**2
    struct[mask] = 1
    return struct


def _thinner_than(n, im):
    return 1 - ndimage.binary_opening(im, structure=_disk(n)).mean()


def granulometry(im, max_size=None, return_pores=False):
    """
    Compute the 'granulometry' of a 2D image by successive morphological
    openings.
    
    Parameters
    ----------
    im : 2D binary array
    
    max_size : int
        maximum size of openings. If None, a value is computed using
        the distance transform.
        
    return_pores : bool
        whether to return the local value of the granulometry. 
        If True, an array of the same shape as im is returned.
        If False, only the histogram of granulometry values is
        returned.
    """
    if max_size is None:
        distance = ndimage.distance_transform_edt(im)
        max_size = int(distance.max())

    if return_pores:
        pores = np.zeros_like(im, dtype=np.uint8)
        for radius in range(1, max_size + 1)[::-1]:
            opened = ndimage.binary_opening(im, structure=_disk(radius))
            mask = np.logical_xor(im, opened)
            pores[mask] = radius
        return pores
    else:
        granulo = Parallel(n_jobs=-1)(delayed(_thinner_than)(n, im)
                                  for n in range(1, max_size + 1))
        return granulo
```

```python
granulo = granulometry(im_bin, return_pores=True)
```

```python
plt.title('granulometry')
plt.imshow(granulo, cmap='Spectral')
plt.colorbar();
```

```python
fig, axs = plt.subplots(1,2, figsize=(15,5))
ax1,ax2=axs

granulo_histo = [x for x in granulo.ravel() if not x == 0];
_ =ax1.hist(granulo_histo, bins=30)
ax1.set_xlabel('local radius')
ax1.set_ylabel('number of pixels')
ax1.title.set_text('Lengthscale from granulometry')

weight=[1./i for i in granulo_histo]
_ =ax2.hist(granulo_histo, weights=weight, bins=30, density=True)
ax2.set_xlabel('local radius')
ax2.set_ylabel('weighted pdf')
ax2.title.set_text('Lengthscale from granulometry')
```
